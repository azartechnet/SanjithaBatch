```python
# Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates whether the data is skewed to the left (negative skewness) or to the right (positive skewness).

# Here's a step-by-step algorithm to calculate and visualize the skewness of the data:

# Step-by-Step Algorithm
# Import Libraries and Load Dataset
# Calculate Skewness
# Interpret Skewness
# Visualize Skewness
```

```python
# Step 1: Import Libraries and Load Dataset
# First, import the necessary libraries and load your dataset. We'll use the titanic dataset for this example.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load dataset
titanic = sns.load_dataset('titanic')

# Display first few rows of the dataset
print(titanic.head())
```

```python
# Step 2: Calculate Skewness
# Calculate the skewness for numerical columns in the dataset.
# Select numerical columns
numerical_columns = titanic.select_dtypes(include=['float64', 'int64']).columns

# Calculate skewness for each numerical column
skewness_values = titanic[numerical_columns].apply(lambda x: skew(x.dropna()))

# Display skewness values
print(skewness_values)
```

```python
# Step 3: Interpret Skewness
# Interpret the skewness values:

# Skewness > 0: Data is positively skewed (right-skewed)
# Skewness < 0: Data is negatively skewed (left-skewed)
# Skewness â‰ˆ 0: Data is symmetric
```

```python
# Step 4: Visualize Skewness
# Visualize the distribution of the numerical columns to understand their skewness.
```

```python
# Plot histograms for numerical columns to visualize skewness
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(titanic[column].dropna(), kde=True)
    plt.title(f'Distribution of {column} (Skewness: {skewness_values[column]:.2f})')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Explanation
# Import Libraries and Load Dataset: We import the required libraries and load the Titanic dataset using Seaborn's load_dataset function.

# Calculate Skewness: We select the numerical columns from the dataset and use the skew function from the scipy.stats module to calculate the skewness for each column.

# Interpret Skewness: The skewness values indicate the asymmetry of the data distribution. Positive values indicate right-skewed data, while negative values indicate left-skewed data.

# Visualize Skewness: We plot histograms with kernel density estimates (KDE) for each numerical column to visualize the distribution and understand the skewness.
```

```python
# Kurtosis is a measure of the "tailedness" of the probability distribution of a real-valued random variable. It helps to understand whether the data distribution is heavy-tailed or light-tailed compared to a normal distribution. Positive kurtosis indicates heavy tails (leptokurtic), while negative kurtosis indicates light tails (platykurtic).

# Here's a step-by-step algorithm to calculate and visualize the kurtosis of the data:

# Step-by-Step Algorithm
# Import Libraries and Load Dataset
# Calculate Kurtosis
# Interpret Kurtosis
# Visualize Kurtosis
```

```python
# Step 1: Import Libraries and Load Dataset
# First, import the necessary libraries and load your dataset. We'll use the titanic dataset for this example.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Load dataset
titanic = sns.load_dataset('titanic')

# Display first few rows of the dataset
print(titanic.head())
```

```python
# Step 2: Calculate Kurtosis
# Calculate the kurtosis for numerical columns in the dataset.
# Select numerical columns
numerical_columns = titanic.select_dtypes(include=['float64', 'int64']).columns

# Calculate kurtosis for each numerical column
kurtosis_values = titanic[numerical_columns].apply(lambda x: kurtosis(x.dropna()))

# Display kurtosis values
print(kurtosis_values)
```

```python
# Step 3: Interpret Kurtosis
# Interpret the kurtosis values:

# Kurtosis > 0: Data has heavy tails (leptokurtic)
# Kurtosis < 0: Data has light tails (platykurtic)
# Kurtosis â‰ˆ 0: Data has normal tails (mesokurtic)
```

```python
# Step 4: Visualize Kurtosis
# Visualize the distribution of the numerical columns to understand their kurtosis.
```

```python
# Step 4: Visualize Kurtosis
# Visualize the distribution of the numerical columns to understand their kurtosis.
# Plot histograms for numerical columns to visualize kurtosis
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(titanic[column].dropna(), kde=True)
    plt.title(f'Distribution of {column} (Kurtosis: {kurtosis_values[column]:.2f})')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Explanation
# Import Libraries and Load Dataset: We import the required libraries and load the Titanic dataset using Seaborn's load_dataset function.

# Calculate Kurtosis: We select the numerical columns from the dataset and use the kurtosis function from the scipy.stats module to calculate the kurtosis for each column.

# Interpret Kurtosis: The kurtosis values indicate the "tailedness" of the data distribution. Positive values indicate heavy tails, while negative values indicate light tails.

# Visualize Kurtosis: We plot histograms with kernel density estimates (KDE) for each numerical column to visualize the distribution and understand the kurtosis.
```

```python
# Probability theory is a fundamental field of mathematics that deals with the likelihood of events occurring. Here are some key concepts and examples of algorithms related to probability theory:

# 1. Basic Probability
# Example: Probability of a single event occurring.

# Steps:
# Define the sample space (all possible outcomes).
# Define the event of interest.
# Calculate the probability of the event.
# Example: Rolling a fair six-sided die
import random

# Sample space
sample_space = [1, 2, 3, 4, 5, 6]

# Event: Rolling a 3
event = [3]

# Probability of the event
probability = len(event) / len(sample_space)
print(f'Probability of rolling a 3: {probability}')
```

```python
# 2. Conditional Probability
# Example: Probability of an event A given that event B has occurred.
#     formula
    
# Steps:
# Define events A and B.
# Calculate the joint probability of A and B.
# Calculate the probability of B.
# Calculate the conditional probability.
```

```python
# Example: Probability of drawing an ace given that the card is a spade from a deck of cards

# Define the deck of cards
deck = [(rank, suit) for rank in range(1, 14) for suit in ['hearts', 'diamonds', 'clubs', 'spades']]

# Event A: Drawing an ace
event_A = [card for card in deck if card[0] == 1]

# Event B: Drawing a spade
event_B = [card for card in deck if card[1] == 'spades']

# Joint event A and B: Drawing an ace of spades
event_A_and_B = [card for card in event_A if card[1] == 'spades']

# Calculate probabilities
P_A_and_B = len(event_A_and_B) / len(deck)
P_B = len(event_B) / len(deck)
P_A_given_B = P_A_and_B / P_B

print(f'Probability of drawing an ace given that the card is a spade: {P_A_given_B}')
```

```python
# Central Limit Theorem
# Example: Demonstrating the central limit theorem with sample means.

# Steps:
# Generate a population with a non-normal distribution.
# Draw multiple random samples from the population.
# Calculate the mean of each sample.
# Plot the distribution of sample means.
```

```python
import numpy as np
# Generate a population with a non-normal distribution (exponential distribution)
population = np.random.exponential(scale=2, size=100000)

# Number of samples and sample size
n_samples = 1000
sample_size = 100

# Draw samples and calculate sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(n_samples)]

# Plot the distribution of sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, kde=True)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Central Limit Theorem: Distribution of Sample Means')
plt.show()
```
