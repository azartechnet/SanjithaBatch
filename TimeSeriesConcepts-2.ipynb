```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file and parse dates
df_fx_data = pd.read_csv('Datasets/SeaPlaneTravel.csv', parse_dates=['Month'])

# Step 2: Set the 'Month' column as the index
df_fx_data.set_index('Month', inplace=True)

# Step 3: Extract the '#Passengers' column as a time series
ts = df_fx_data['#Passengers']

# Step 4: Resample to daily frequency (assuming interpolation for missing days)
ts_daily = ts.resample('D').interpolate(method='linear')  # Linear interpolation for missing values

# Step 5: Plot the time series data daywise
plt.figure(figsize=(15, 6))
plt.plot(ts_daily, label='#Passengers')
plt.xlabel('Date')
plt.ylabel('#Passengers')
plt.title('Daily #Passengers Over Time')
plt.legend()
plt.show()
```
