```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualization settings
plt.style.use('ggplot')
sns.set(style='whitegrid')
```

```python
# Load dataset
data = pd.read_csv('Datasets/customer_churn.csv')

# Display the first few rows
print(data.head())

# Display basic information about the dataset
print(data.info())

# Display summary statistics
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())

# Handle missing values by dropping rows with missing values
data = data.dropna()

# Alternatively, fill missing values
# data['column_name'] = data['column_name'].fillna(data['column_name'].mean())

# Convert categorical variables to numeric
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    data[column] = data[column].astype('category').cat.codes
```

```python
# Distribution of numerical features
data.hist(bins=50, figsize=(20, 15))
plt.show()

# Distribution of categorical features
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=column, data=data.head())
    plt.show()
```

```python
# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

# Relationships between numerical features and target variable
target_variable = 'Churn'  # Replace with your target variable name
sns.pairplot(data, hue=target_variable, palette='coolwarm')
plt.show()

# Relationships between categorical features and target variable
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=column, y=target_variable, data=data)
    plt.show()
```
