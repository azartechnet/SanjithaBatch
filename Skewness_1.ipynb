```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
#Load dataset
titanic=sns.load_dataset('titanic')

print(titanic.head())
```

```python
#Step 2 Calculate Skewness
#calculate the skewness for numerical columns in the dataset
#selete numerical columns

numerical_columns=titanic.select_dtypes(include=['float64','int64']).columns;
```

```python
numerical_columns
```

```python
#calculate skewness for each numbericak columns;
skewness_values=titanic[numerical_columns].apply(lambda x: skew(x.dropna()))

print(skewness_values)
```

```python
# Step 3: Interpret Skewness
# Interpret the skewness values:

# Skewness > 0: Data is positively skewed (right-skewed)
# Skewness < 0: Data is negatively skewed (left-skewed)
# Skewness â‰ˆ 0: Data is symmetric
```

```python
#Step 4 visualize skewness
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10,6))
    sns.histplot(titanic[column].dropna(),kde=True)
    plt.title("Distribution")
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
```

```python
#Load dataset
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#Step 2 Calculate kurtosis
numerical_columns=titanic.select_dtypes(include=['float64','int64']).columns
```

```python
numerical_columns
```

```python
kurtosis_values=titanic[numerical_columns].apply(lambda x:kurtosis(x.dropna()))
```

```python
kurtosis_values
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10,6))
    sns.histplot(titanic[column].dropna(),kde=True)
    plt.title("Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Example: Probability of drawing an ace given that the card is a spade from a deck of cards

# Define the deck of cards
deck = [(rank, suit) for rank in range(1, 14) for suit in ['hearts', 'diamonds', 'clubs', 'spades']]

# Event A: Drawing an ace
event_A = [card for card in deck if card[0] == 1]

# Event B: Drawing a spade
event_B = [card for card in deck if card[1] == 'spades']

# Joint event A and B: Drawing an ace of spades
event_A_and_B = [card for card in event_A if card[1] == 'spades']

# Calculate probabilities
P_A_and_B = len(event_A_and_B) / len(deck)
P_B = len(event_B) / len(deck)
P_A_given_B = P_A_and_B / P_B

print(f'Probability of drawing an ace given that the card is a spade: {P_A_given_B}')
```
