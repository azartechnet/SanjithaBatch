```python
# 1. Normal Distribution
# The normal distribution, also known as the Gaussian distribution, is a continuous 
#probability distribution characterized by a bell-shaped curve. Here's an example of 
#how to generate and visualize a normal distribution:
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the normal distribution
mu, sigma = 0, 1  # mean and standard deviation

# Generate a sample of 1000 data points
data = np.random.normal(mu, sigma, 1000)

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30)
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Z-score
# A Z-score represents the number of standard deviations a data point is from the mean of the distribution. The formula for the Z-score is:
```

```python
# Generate a dataset
data = np.random.normal(50, 15, 1000)  # mean=50, std=15

# Calculate the mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Calculate the Z-scores
z_scores = [(x - mean) / std for x in data]

# Display the first 10 Z-scores
print(z_scores[:10])
```

```python
# 3. Confidence Interval
# A confidence interval gives an estimated range of values which is likely to include an unknown population parameter. The formula for a 95% confidence interval for the mean is:
```

```python
import scipy.stats as stats

# Generate a dataset
data = np.random.normal(50, 15, 1000)  # mean=50, std=15

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Calculate the Z-value for a 95% confidence interval
z_value = stats.norm.ppf(0.975)  # 0.975 corresponds to 95% CI

# Calculate the margin of error
margin_of_error = z_value * (sample_std / np.sqrt(n))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f'95% Confidence Interval: {confidence_interval}')
```

```python
# Explanation
# Normal Distribution: Generates a sample dataset from a normal distribution and visualizes it using a histogram with a KDE plot.
# Z-score: Calculates the Z-scores for the dataset, indicating how many standard deviations each data point is from the mean.
# Confidence Interval: Computes the 95% confidence interval for the sample mean, providing a range that is likely to contain the true population mean.
```

```python
#Z-distribution
import numpy as np
import scipy.stats as stats

# Population parameters
population_mean = 70
population_std = 10

# Sample data
sample_size = 30
sample_mean = 73

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))  # Two-tailed test

# Determine the critical value for a 95% confidence level (alpha = 0.05)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test

# Print the results
print(f'Z-score: {z_score}')
print(f'P-value: {p_value}')
print(f'Critical value: {critical_value}')

# Decision
if np.abs(z_score) > critical_value:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
