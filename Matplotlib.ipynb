```python
x_data=[1,2,3,4,5]
y1_data=[10,15,13,17,10]
y2_data=[8,13,11,15,8]

#create a stacked area chart

fig=px.area(x=x_data,y=[y1_data,y2_data],stacked=True,labels={'Y':'Y-axis'})

#Update layout

fig.update_layout(title='Stacked Area chart',xaxis_title='X-axis')

fig.show()
```

```python
import plotly.graph_objects as go

# Sample data
x_data = [1, 2, 3, 4, 5]
y1_data = [1, 3, 2, 4, 3]
y2_data = [8, 13, 11, 15, 8]

fig=go.Figure()

fig.add_trace(go.scatter(x=x_data,y=y1_data,stackgroup='One',name='Trace1'))

fig.add_trace(go.scatter(x=x_data,y=y2_data,stackgroup='two',name='Trace2'))

fig.update_layout(title='Stacked area chart',xaxis_title='xaxis')

fig.show()
```
