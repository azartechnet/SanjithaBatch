```python
#Polynomial Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.DataCollection
```

```python
data=pd.read_csv("Datasets/Position_Salaries.csv")
data
```

```python
#Separate the input and output
X=data.iloc[:,1:2].values
Y=data.iloc[:,2].values
```

```python
Y
```

```python
#3.DataAnlsis
plt.plot(X,Y)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,Y)
```

```python
#Fitting and training model
predicated_value=lin_reg.predict(X)
```

```python
#predicated_value
```

```python
#Visualizing the linear regression result
plt.scatter(X,Y,color='red')
plt.plot(X,predicated_value,color='blue')
plt.title("Truth or bluff Linear Regression")
plt.xlabel("PosistionLevel")
plt.ylabel("Salary")
plt.show()
```

```python
#fitting polynomial Regression  to the dataset
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol_reg=PolynomialFeatures(degree=4)
X_poly=pol_reg.fit_transform(X)
```

```python
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,Y)
predicated_value=lin_reg2.predict(pol_reg.fit_transform(X))
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,predicated_value,color='blue')
plt.title("Truth")
plt.xlabel("PosistionLevel")
plt.ylabel("Salary")
plt.show()
```
